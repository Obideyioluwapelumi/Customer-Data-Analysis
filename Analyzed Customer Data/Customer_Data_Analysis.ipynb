### Age group with Most Profession 

```python
plt.figure(figsize=(10,10))
sns.barplot(x = "Profession",y="Age", data = customers, palette="Spectral")
plt.title("Age Vs Profession", fontsize=30)
plt.xlabel("Profession", fontsize=20)
plt.ylabel("Age", fontsize=20)
plt.xticks(rotation = "vertical")
plt.show()
```

### Number of Male and Female

```python
plt.figure(figsize=(10,10))
sns.countplot(x=customers["Gender"],palette="flare")
plt.title("Number of Male and Female",fontsize=30)
plt.xlabel("Gender",fontsize=20)
plt.ylabel("Count",fontsize=20)
plt.show()
```

### Profession gettign the highest Income

```python
plt.figure(figsize=(20,15))
sns.barplot(x="Profession",y="Annual Income ($)",data=customers)
plt.title("Professions and its Annual Income",fontsize=35)
plt.xlabel("Profession",fontsize=25)
plt.ylabel("Annual Income",fontsize=25)
plt.xticks(rotation="vertical",fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

### Distribution of Work Experience

```python
customers["Work Experience"]=customers["Work Experience"].astype("str")
```

```python
plt.figure(figsize=(15,10))
sns.histplot(data=customers, x= "Work Experience")
plt.xlabel('Work Experience',fontsize=15)
plt.ylabel("count",fontsize=15)
plt.title("Distribution of Work Experience",fontsize=30)
plt.show()
```

### Profession chosen by most people
